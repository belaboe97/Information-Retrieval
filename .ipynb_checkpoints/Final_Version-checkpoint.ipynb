## Preprocessing

```python
stemmer = WordNetLemmatizer()
sw_nltk = stopwords.words('english')
#preprocess
def preprocess(sen : str): 
    # Remove all the special characters
    document = re.sub(r'\W', ' ', sen)
    
    # remove all single characters
    document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
    
    # Remove single characters from the start
    document = re.sub(r'\^[a-zA-Z]\s+', ' ', document) 
    
    # Substituting multiple spaces with single space
    document = re.sub(r'\s+', ' ', document, flags=re.I)
    
    # Removing prefixed 'b'
    document = re.sub(r'^b\s+', '', document)

    # Converting to Lowercase
    document = document.lower()
    
    # Lemmatization
    document = document.split()

    document = [stemmer.lemmatize(word) for word in document]
    document = ' '.join(document)
    """
    document = [word for word in document.split() if word.lower() not in sw_nltk]
    document = " ".join(document)
    """
    return document

# Remove unecessary parts of the articles and word stemm the words for better accuracy
def preprocess_articles(df): 
  documents = []
  for sen in range(0, len(df.text)):
    document = preprocess(str(df.text[sen]))
    documents.append(document)
  result = df.join(pd.DataFrame(documents, columns=["preprocessed_text"]))
  return result

def preprocess_query(query : str): 
  return preprocess(query)

df = preprocess_articles(df)
```

#### Functions related to build the specific Index

```python
#Code obtained and adjusted from 

# compute Term frequency of a specific term in a document
# https://t4tutorials.com/cosine-similarity-in-data-mining/
def termFrequency(term, document):
  return document.count(term.lower()) / float(len(document))

# IDF of a term
def inverseDocumentFrequency(term, documents):
  count = 0
  for doc in documents:
      if term.lower() in doc:
          count += 1
  if count > 0:
      return  1 + math.log(float(len(documents))/count)
  else:
      return 1.0
```

#### Functions related to build the specific Index for given Query

```python
def generateVectors(query, documents):
    tf_idf_matrix = np.zeros((len(query.split()), len(documents)))
    for i, s in enumerate(query.lower().split()):
        idf = inverseDocumentFrequency(s, documents)
        for j,doc in enumerate(documents):
            tf_idf_matrix[i][j] = idf * termFrequency(s, doc)
    return tf_idf_matrix
    

def word_count(s):
    counts = dict()
    words = s.lower().split()
    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    return counts

def build_query_vector(query, documents):
    count = word_count(query)
    vector = np.zeros((len(count),1))
    for i, word in enumerate(query.lower().split()):
        vector[i] = float(count[word])/len(count) * inverseDocumentFrequency(word, documents)
    return vector
```

#### Fucntions related to compare query and tfidf matrix for similiarity

```python
#Generate Matrix and query vector
# https://ted-mei.medium.com/demystify-tf-idf-in-indexing-and-ranking-5c3ae88c3fa0
def generate_matrix_and_query(query, documents):
  #query = preprocess_articles(query)
  tf_idf_matrix = generateVectors(query, documents)
  query_vector = build_query_vector(query, documents)
  return  query_vector, tf_idf_matrix

#Calcuate cosine similiarity
def cosine_similarity(v1, v2):
    return np.dot(v1,v2)/float(LA.norm(v1)*LA.norm(v2))
    
def compute_relevance_m1(query, documents,  query_vector, tf_idf_matrix):
    help_arr = []
    for i, doc in enumerate(documents):
        similarity = cosine_similarity(tf_idf_matrix[:,i].reshape(1, len(tf_idf_matrix)), query_vector)
        #print("query document {}, similarity {}".format(i, float(similarity[0])))
        sim = float(similarity[0])
        #print(sim, math.isnan(sim))
        if not math.isnan(sim): 
          help_arr.append((i,sim))
    return sorted(help_arr,key=lambda tup: tup[1],reverse=True)
```
